# Q1 Write a python program which searches all the product under a particular product from www.amazon.in. The
product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for
guitars. 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
url = "https://www.amazon.in/"
driver.get(url)
time.sleep(3) 

try:
  
    product_name = input("Enter the product to search: ")

  
    search_box = driver.find_element(By.ID, "twotabsearchtextbox")
    search_box.clear()
    search_box.send_keys(product_name)

    search_button = driver.find_element(By.XPATH, '//input[@value="Go"]')
    search_button.click()
    time.sleep(3) 

    
    product_elements = driver.find_elements(By.XPATH, '//div[@data-component-type="s-search-result"]')

    for product in product_elements:
        try:
            product_name = product.find_element(By.XPATH, './/h2/a/span').text
            product_price = product.find_element(By.XPATH, './/span[@class="a-price-whole"]').text
            product_rating = product.find_element(By.XPATH, './/span[@class="a-icon-alt"]').get_attribute('innerHTML')

           
            print(f"Product Name: {product_name}")
            print(f"Price: ₹{product_price}")
            print(f"Rating: {product_rating}")
            print("="*50)

        except Exception as e:
            print("Error processing product:", e)

except Exception as e:
    print("An error occurred:", str(e))

finally:
  
    driver.quit()
```

# 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search
results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then
scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. 

```python


def scrape_amazon_products(search_query):

    driver = webdriver.Chrome()
    driver.maximize_window()


    driver.get("https://www.amazon.in/")

    try:
  
        search_box = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "twotabsearchtextbox"))
        )
        
   
        search_box.clear()
        search_box.send_keys(search_query)
        search_box.send_keys(Keys.RETURN)

       
        brand_names = []
        product_names = []
        prices = []
        return_exchange = []
        expected_delivery = []
        availability = []
        product_urls = []

        page_number = 1
        while page_number <= 3:
     
            results = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div.s-result-item"))
            )

    
            for result in results:
                try:
        
                    brand_name = result.find_element(By.CSS_SELECTOR, "span.a-size-base-plus").text.strip()
                except:
                    brand_name = "-"
                
                try:
                    product_name = result.find_element(By.CSS_SELECTOR, "span.a-text-normal").text.strip()
                except:
                    product_name = "-"
                
                try:
                    price = result.find_element(By.CSS_SELECTOR, "span.a-price").text.strip()
                except:
                    price = "-"
                
                try:
                    return_exchange_info = result.find_element(By.CSS_SELECTOR, "div.a-row.a-size-small span:nth-child(2)").text.strip()
                except:
                    return_exchange_info = "-"
                
                try:
                    expected_delivery_info = result.find_element(By.CSS_SELECTOR, "div.a-row.a-size-small span:nth-child(4)").text.strip()
                except:
                    expected_delivery_info = "-"
                
                try:
                    availability_info = result.find_element(By.CSS_SELECTOR, "div.a-row.a-size-small span.a-size-base").text.strip()
                except:
                    availability_info = "-"
                
                try:
                    product_url = result.find_element(By.CSS_SELECTOR, "a.a-link-normal").get_attribute('href')
                except:
                    product_url = "-"

               
                brand_names.append(brand_name)
                product_names.append(product_name)
                prices.append(price)
                return_exchange.append(return_exchange_info)
                expected_delivery.append(expected_delivery_info)
                availability.append(availability_info)
                product_urls.append(product_url)
            
         
            try:
                next_page = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.CSS_SELECTOR, "li.a-last a"))
                )
                next_page.click()
                page_number += 1
                time.sleep(3) 
            except:
                break 
        df = pd.DataFrame({
            "Brand Name": brand_names,
            "Name of the Product": product_names,
            "Price": prices,
            "Return/Exchange": return_exchange,
            "Expected Delivery": expected_delivery,
            "Availability": availability,
            "Product URL": product_urls
        })

      
        df.to_csv(f"{search_query}_products.csv", index=False)
        print(f"Scraping and saving of {search_query} products completed!")

    finally:
        driver.quit()


search_query = input("Enter the product you want to search on Amazon.in: ")
scrape_amazon_products(search_query)
```

# Q.3 Write a python program to access the search bar and search button on images.google.com and scrape 10
images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.

```python
import os
import time
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By


keywords = ['fruits', 'cars', 'Machine Learning', 'Guitar', 'Cakes']
num_images_per_keyword = 10


driver = webdriver.Chrome()
driver.maximize_window()

for keyword in keywords:
    print(f"Fetching images for '{keyword}'...")

   
    driver.get("https://images.google.com/")
    time.sleep(2) 

    
    search_box = driver.find_element(By.XPATH,'//*[@id="APjFqb"]')
    search_box.clear()
    search_box.send_keys(keyword)
    search_box.send_keys(Keys.RETURN)
    time.sleep(2) 
 
    for _ in range(5):  #
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)

    
    thumbnail_results = driver.find_elements(By.XPATH, "//img[@class='Q4LuWd']")
    image_urls = set()

    
    for img in thumbnail_results[:num_images_per_keyword]:
        try:
            img.click()
            time.sleep(1) 
            actual_images = driver.find_elements(By.XPATH, "//img[@class='n3VNCb']")
            for actual_image in actual_images:
                src = actual_image.get_attribute('src')
                if src and 'http' in src:
                    image_urls.add(src)
        except Exception as e:
            print(f"Error: {str(e)}")

   
    save_directory = os.path.join(".", keyword)
    os.makedirs(save_directory, exist_ok=True)

    for i, image_url in enumerate(image_urls):
        try:
            response = requests.get(image_url)
            response.raise_for_status() 
            with open(os.path.join(save_directory, f"{keyword}_{i+1}.jpg"), "wb") as f:
                f.write(response.content)
        except Exception as e:
            print(f"Error downloading {image_url}: {str(e)}")

    print(f"Image scraping for '{keyword}' completed!")


driver.quit()

print("Image scraping completed!")

```

# Q.4 Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com

Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”,
“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the
details is missing then replace it by “- “. Save your results in a dataframe and CSV. 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Set up Selenium WebDriver
driver = webdriver.Chrome()

# List of smartphone models to search
smartphone_models = ['Oneplus Nord']  # Add more models as needed

# DataFrame to store results
df = pd.DataFrame(columns=["Brand Name", "Smartphone name", "Colour", "RAM", "Storage(ROM)",
                           "Primary Camera", "Secondary Camera", "Display Size", "Battery Capacity",
                           "Price", "Product URL"])

for model in smartphone_models:
    driver.get("https://www.flipkart.com/")
    
  

    
    search_box = driver.find_element(By.NAME, "q")
    search_box.clear()
    search_box.send_keys(model)
    search_box.send_keys(u'\ue007')  
    
    time.sleep(2)  
 
    product_elements = driver.find_elements(By.XPATH, '//a[@class="CGtC98"]')
    product_urls = [elem.get_attribute('href') for elem in product_elements]
    
    for url in product_urls:
        driver.get(url)
        time.sleep(2)
        
      
        try:
            brand_name = driver.find_element(By.XPATH, '').text.split()[0]
        except:
            brand_name = "-"
        
        try:
            smartphone_name = driver.find_element(By.XPATH, '//span[@class="B_NuCI"]').text
        except:
            smartphone_name = "-"
        
        try:
            colour = driver.find_element(By.XPATH, '//td[text()="Color"]/following-sibling::td//li').text
        except:
            colour = "-"
        
        try:
            ram = driver.find_element(By.XPATH, '//td[text()="RAM"]/following-sibling::td//li').text
        except:
            ram = "-"
        
        try:
            storage = driver.find_element(By.XPATH, '//td[text()="Internal Storage"]/following-sibling::td//li').text
        except:
            storage = "-"
        
        try:
            primary_camera = driver.find_element(By.XPATH, '//td[text()="Primary Camera"]/following-sibling::td//li').text
        except:
            primary_camera = "-"
        
        try:
            secondary_camera = driver.find_element(By.XPATH, '//td[text()="Secondary Camera"]/following-sibling::td//li').text
        except:
            secondary_camera = "-"
        
        try:
            display_size = driver.find_element(By.XPATH, '//*[@id="container"]/div/div[3]/div[1]/div[2]/div[8]/div[5]/div/div[2]/div[1]/div[2]/table/tbody/tr[1]/td[2]/ul/li').text
        except:
            display_size = "-"
        
        try:
            battery_capacity = driver.find_element(By.XPATH, '//td[text()="Battery Capacity"]/following-sibling::td//li').text
        except:
            battery_capacity = "-"
        
        try:
            price = driver.find_element(By.XPATH, '//div[@class="_30jeq3 _16Jk6d"]').text
        except:
            price = "-"
        
        product_url = url
        
       
        df = df.append({
            "Brand Name": brand_name,
            "Smartphone name": smartphone_name,
            "Colour": colour,
            "RAM": ram,
            "Storage(ROM)": storage,
            "Primary Camera": primary_camera,
            "Secondary Camera": secondary_camera,
            "Display Size": display_size,
            "Battery Capacity": battery_capacity,
            "Price": price,
            "Product URL": product_url
        }, ignore_index=True)


df.to_csv("smartphones.csv", index=False)


driver.quit()
```

# Q5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps. 


```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import re

city_name = "Pune"

driver = webdriver.Chrome()

driver.get("https://www.google.com/maps")
time.sleep(2)

search_box = driver.find_element(By.XPATH, '//input[@id="searchboxinput"]')
search_box.send_keys(city_name)
search_box.send_keys(u'\ue007')
time.sleep(3)

current_url = driver.current_url

match = re.search(r'@(-?\d+\.\d+),(-?\d+\.\d+)', current_url)
if match:
    latitude = match.group(1)
    longitude = match.group(2)
    print(f"The coordinates of {city_name} are: Latitude = {latitude}, Longitude = {longitude}")
else:
    print(f"Could not find coordinates for {city_name}.")

driver.quit()
```

# Q6. Write a program to scrap all the available details of best gaming laptops from digit.in

```python

from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

driver = webdriver.Chrome()
url = "https://www.digit.in/top-products/best-gaming-laptops-40.html"
driver.get(url)
time.sleep(5)

laptop_containers = driver.find_elements(By.XPATH,'//div[@class="r_offer_details news-community clearfix  product type-product"]')

laptops = []

for container in laptop_containers:
    laptop = {}
    
    try:
        laptop['Name'] = container.find_element(By.XPATH,'//h3[@class="font130 mt0 mb10 mobilesblockdisplay "]').text.strip()
    except:
        laptop['Name'] = '-'
    
    try:
        specs = container.find_elements(By.XPATH, '//div[@class="woo_code_zone_loop clearbox"]')
        for spec in specs:
            spec_title = spec.find_element(By.CLASS_NAME, '').text.strip()
            spec_value = spec.find_element(By.CLASS_NAME, 'Spcs-value').text.strip()
            laptop[spec_title] = spec_value
    except:
        pass
    
    laptops.append(laptop)

driver.quit()

df = pd.DataFrame(laptops)
df.to_csv('best_gaming_laptops.csv', index=False)
print(df)
```

# Q.7 Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped:
“Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”. 


```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

driver = webdriver.Chrome()

url = "https://www.forbes.com/billionaires/"
driver.get(url)
time.sleep(5)

billionaires = []

for row in url:
    billionaire = {}
    try:
        billionaire['Rank'] = row.find_element(By.XPATH, './/div[@class="rank"]').text.strip()
    except:
        billionaire['Rank'] = '-'
    
    try:
        billionaire['Name'] = row.find_element(By.XPATH, 'class="Table_personName__Bus2E"').text()
    except:
        billionaire['Name'] = '-'
    
    try:
        billionaire['Net worth'] = row.find_element(By.XPATH, './/div[@class="netWorth"]').text.strip()
    except:
        billionaire['Net worth'] = '-'
    
    try:
        billionaire['Age'] = row.find_element(By.XPATH, './/div[@class="age"]').text.strip()
    except:
        billionaire['Age'] = '-'
    
    try:
        billionaire['Citizenship'] = row.find_element(By.XPATH, './/div[@class="countryOfCitizenship"]').text.strip()
    except:
        billionaire['Citizenship'] = '-'
    
    try:
        billionaire['Source'] = row.find_element(By.XPATH, './/div[@class="source"]').text.strip()
    except:
        billionaire['Source'] = '-'
    
    try:
        billionaire['Industry'] = row.find_element(By.XPATH, './/div[@class="category"]').text.strip()
    except:
        billionaire['Industry'] = '-'
    
    billionaires.append(billionaire)

driver.quit()

df = pd.DataFrame(billionaires)
df.to_csv('forbes_billionaires.csv', index=False)
print(df)
```

# Q8. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted
from any YouTube Video

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd


driver = webdriver.Chrome()


video_url = 'https://www.youtube.com/watch?v=pZbsVynV2so' 
driver.get(video_url)
time.sleep(5)  


for _ in range(20): 
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(3) 

comments_data = []


comments_section = driver.find_element(By.XPATH, '//*[@id="sections"]')
comments = comments_section.find_elements(By.XPATH, '//*[@id="content-text"]')
upvotes = comments_section.find_elements(By.XPATH, '//*[@id="vote-count-middle"]')
timestamps = comments_section.find_elements(By.XPATH, '//*[@id="header-author"]/yt-formatted-string/a')

for i in range(min(len(comments), 500)): 
    comment = comments[i].text
    upvote = upvotes[i].text if i < len(upvotes) else "0"
    timestamp = timestamps[i].text if i < len(timestamps) else "N/A"
    comments_data.append({
        'comment': comment,
        'upvotes': upvote,
        'published_at': timestamp
    })


driver.quit()

df = pd.DataFrame(comments_data)
df.to_csv('youtube_comments.csv', index=False)
print(df)
```

# 9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in
“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall
reviews, privates from price, dorms from price, facilities and property description.

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException

def scrape_hostels_in_london():
    driver = webdriver.Chrome()
    driver.maximize_window()

    try:
        url = "https://www.hostelworld.com/pwa/wds/s?q=London%2C+England&country=England&city=London&type=city&id=3&from=2024-06-14&to=2024-06-17&guests=1&page=1"
        driver.get(url)
        
        
        
        hostel_list = []

        hostels = driver.find_elements(By.XPATH, '//div[@class="property-info-container"]')
        for hostel in hostels:
            try:
                name = hostel.find_elements(By.XPATH, '//*[@id="__layout"]/div/div[2]/div[2]/div[1]/div/div/div[5]/div[1]/a/a/div[2]/div[1]/div[2]/span').text.strip()
            except NoSuchElementException:
                name = "-"

            try:
                distance = hostel.find_elements(By.XPATH, ".//span[contains(@class, 'distFromCenter')]").text.strip()
            except NoSuchElementException:
                distance = "-"

            try:
                ratings = hostel.find_element(By.XPATH, ".//div[contains(@class, 'rating')]//span[contains(@class, 'score')]").text.strip()
            except NoSuchElementException:
                ratings = "-"

            try:
                total_reviews = hostel.find_element(By.XPATH, ".//div[contains(@class, 'review')]//span[contains(@class, 'total')]").text.strip()
            except NoSuchElementException:
                total_reviews = "-"

            try:
                overall_reviews = hostel.find_element(By.XPATH, ".//div[contains(@class, 'rating')]//span[contains(@class, 'label')]").text.strip()
            except NoSuchElementException:
                overall_reviews = "-"

            try:
                privates_from_price = hostel.find_element(By.XPATH, ".//div[contains(@class, 'private')]//span[contains(@class, 'price')]").text.strip()
            except NoSuchElementException:
                privates_from_price = "-"

            try:
                dorms_from_price = hostel.find_element(By.XPATH, ".//div[contains(@class, 'dorms')]//span[contains(@class, 'price')]").text.strip()
            except NoSuchElementException:
                dorms_from_price = "-"

            try:
                facilities = hostel.find_element(By.XPATH, ".//ul[contains(@class, 'facilities')]").text.strip()
            except NoSuchElementException:
                facilities = "-"

            try:
                description = hostel.find_element(By.XPATH, ".//div[contains(@class, 'description')]//p").text.strip()
            except NoSuchElementException:
                description = "-"

            hostel_data = {
                "Name": name,
                "Distance from City Centre": distance,
                "Ratings": ratings,
                "Total Reviews": total_reviews,
                "Overall Reviews": overall_reviews,
                "Privates From Price": privates_from_price,
                "Dorms From Price": dorms_from_price,
                "Facilities": facilities,
                "Description": description
            }

            hostel_list.append(hostel_data)

       
        print("data saved")

    except Exception as e:
        print(f"Error: {str(e)}")
    finally:
        driver.quit()

scrape_hostels_in_london()
```
